# Dimensions for Sticking Probabilities

```python
from boxcount import *

figure_folder = "../Report/Figures/"
```

```python
p = 1.0

sim = DLA(5000, 20000, p)
d = sim.boxcount_dimension(1, 20, 1, True)
print(f"p={p}:\t {d[0]:.3f} ± {d[1]:.3f}")
```

```python
p = 0.5

# sim = DLA(5000, 20000, 1.0)
simulation = DLA(15000, 50000, p)

ds = sim.boxcount_dimension(1, 100, 1, True)
d = simulation.boxcount_dimension(1, 100, 5, True)

print(f"p={p}:\t {d[0]:.3f} ± {d[1]:.3f}")
print(f"p=1.0:\t {ds[0]:.3f} ± {ds[1]:.3f}")
```

```python
p = 0.1

# sim = DLA(5000, 20000, 1.0)
simulation = DLA(17000, 100000, p)

ds = sim.boxcount_dimension(1, 100, 1, True)
d = simulation.boxcount_dimension(1, 100, 5, True)

print(f"p={p}:\t {d[0]:.3f} ± {d[1]:.3f}")
print(f"p=1.0:\t {ds[0]:.3f} ± {ds[1]:.3f}")
```

```python
p = 0.05

# sim = DLA(5000, 20000, 1.0)
simulation = DLA(20000, 220000, p)

ds = sim.boxcount_dimension(1, 100, 1, True)
d = simulation.boxcount_dimension(1, 100, 5, True)

print(f"p={p}:\t {1.71*d[0]/ds[0]:.3f} ± {d[1]:.3f}")
print(f"p=1.0:\t {ds[0]:.3f} ± {ds[1]:.3f}")
```

```python
p = 0.01

# sim = DLA(5000, 20000, 1.0)
simulation = DLA(21000, 250000, p)

ds = sim.boxcount_dimension(1, 100, 1, True)
d = simulation.boxcount_dimension(1, 100, 5, True)

print(f"p={p}:\t {1.71*d[0]/ds[0]:.3f} ± {d[1]:.3f}")
print(f"p=1.0:\t {ds[0]:.3f} ± {ds[1]:.3f}")
```

```python
sim1 = DLA(10000, 50000, 1.0) #15000, 50000
sim2 = DLA(10000, 50000, 0.01)
```

```python
fig= plt.figure(figsize=[8, 8])

plt.scatter(sim1.data['x'], sim1.data['y'], label="p=1", marker="s", color="black", s=1)
plt.scatter(sim2.data['x']-150, sim2.data['y']-150, label="p=0.05", marker="s", color="red", s=1)
plt.legend()

plt.xlim([0, 500])
plt.ylim([0, 500])
plt.axis('square')
# plt.axis('off')

plt.show()
```

```python
fig.savefig(figure_folder + "3.png", dpi=300, transparent=True, padding=0)
```
