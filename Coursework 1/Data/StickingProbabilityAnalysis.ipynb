# Dimensions for Sticking Probabilities

```python
from boxcount import *

p = 1.0

sim = DLA(5000, 20000, p)
d = sim.boxcount_dimension(1, 20, 1, True)
print(f"p={p}:\t {d[0]:.3f} Â± {d[1]:.3f}")
```

```python
p = 0.5

# sim = DLA(5000, 20000, 1.0)
simulation = DLA(15000, 50000, p)

ds = sim.boxcount_dimension(1, 100, 1, True)
d = simulation.boxcount_dimension(1, 100, 5, True)

print(f"p={p}:\t {d[0]:.3f} Â± {d[1]:.3f}")
print(f"p=1.0:\t {ds[0]:.3f} Â± {ds[1]:.3f}")
```
