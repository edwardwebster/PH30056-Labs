# Extra Figures

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from boxcount import *
figure_folder = "../Report/Figures/"
```

```python
data = pd.read_csv("Extra/multiple_seeds.csv")
data.head()

cluster = data.state == 1
walker = data.state == -1

fig = plt.figure(figsize=[8,8])

plt.scatter(data['x'][cluster], data['y'][cluster], s=1, color="red", label="Cluster Particles")
plt.scatter(data['x'][walker], data['y'][walker], s=0.05, color="black", label="Walker Particles")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.axis('square')

plt.show()

fig.savefig(figure_folder + "multi_seeds.png", dpi=300, transparent=True, padding=0)
```

```python
fractal_dimension("Extra/multiple_seeds.csv", 2, 70, 1, True, False)
```
