```python
data1 = pd.read_csv(f"PH30056/dla.csv")
data1.head()

plt.plot(data1["n"], data1["r"])
```

```python
x = []
y = []

i = 10
while i>0:
    try:
        data = pd.read_csv(f"{folder}/output.csv.{i}")
        x.append(i)
        y.append(max(abs(data['r'])))
    except:
        print(i)
        break
    i += 10;

# x = np.log(x)
# y = np.log(y)
plt.figure()
plt.plot(data1["n"], data1["r"], label="Original")
plt.plot(x, y, label="Mine")
plt.legend()
plt.show()
[m, c] = np.polyfit(x, y, 1)
# plt.plot(x, m*x + c, color="red")
print(m)
```

```python
plt.plot(np.log(data1["n"]), np.log(data1["r"]), label="Original")
plt.plot(np.log(x), np.log(y), label="Mine")
plt.legend()

[m, c] = np.polyfit(np.log(x), np.log(y), 1)
[m2, _] = np.polyfit(np.log(data1["n"]), np.log(data1["r"]), 1)
print(m)
print(m2)
```
